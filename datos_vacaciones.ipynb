```python
#Se usan graficos de dispersión para una verificación visual rápida de la naturaleza de la relación entre las variables.
#En esta practica de laboratorio se usa el coeficiente de correlacion de pearson
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Para asegurar que los datos no sean sesgados se dividirá la estructura en dos
#La estructura de los hombres y la de las mujeres


menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

```python
#El dataframe incluye 3 medidas de inteligencia PIQ,FSIQ,VIQ
#Se usa el metodo mean para sacar el promedio, se usa el scatter para hacer el diagrama de dispersión
menMeanSmarts = menDf[['PIQ','FSIQ','VIQ']].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
womenMeanSmarts = womenDf[['PIQ','FSIQ','VIQ']].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
#El metodo corr es para calcular la correlación segúnb una estructura de datos
brainFrame.corr(method='pearson')
```
