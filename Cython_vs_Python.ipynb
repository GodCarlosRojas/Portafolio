# Metodos Para Resultados
<BLOCKQUOTE>Para realizar un correcto analisis del laboratorio, se opto por utilizar las herramienteas de python para graficar en un orden adecuado, a continuacion el algoritmo implementado en este trabajo.</BLOCKQUOTE>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def graficar(py,cy):
# La primer linea de graficas corresponde a las 3 pruebas  realizadas en python
  for i in range(1,4,1):

    plt.subplot(3,3,i)
    plt.plot(py[i-1],label='Phyton')
    plt.axhline(y=py[i-4].mean(), xmin=0, xmax=29, color='red',label='Promedio')
    plt.xlabel('Iteracion')
    plt.ylabel('Tiempo (s)')
    plt.legend()
    plt.grid()
    plt.title('Prueba {}'.format(i))

#La segunda linea de graficas corresponde a las 3 pruebas realizadas en Cython
  for i in range(4,7,1):

    plt.subplot(3,3,i)
    plt.plot(cy[i-4],color='orange',label='Cython')
    plt.axhline(y=cy[i-4].mean(), xmin=0, xmax=29, color='green',label='Promedio')
    plt.xlabel('Iteracion')
    plt.ylabel('Tiempo (s)')
    plt.legend()
    plt.grid()
    plt.title('Prueba {}'.format(i-3))

# La tercer linea de graficas corresponde a la comparativa de resultados obtenidos por python
# contras los obtenidos en Cython
  for i in range(7,10,1):

    plt.subplot(3,3,i)
    plt.plot(py[i-7],label='Python')
    plt.plot(cy[i-7],color='orange',label='Cython')
    plt.xlabel('Iteracion')
    plt.ylabel('Tiempo (s)')
    plt.legend()
    plt.grid()
    plt.title('Prueba {}'.format(i-6))

  plt.tight_layout()
```
