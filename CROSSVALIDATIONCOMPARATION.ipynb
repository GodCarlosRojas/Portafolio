```python
df = pd.DataFrame([[2,"si"],
                   [4,"si"],
                   [6,"no"],
                   [9,"no"],
                   [10,"no"],
                   [11,"si"],
                   [15,"si"],
                   [20,"no"],
                   [21,"si"]
                   ],columns=["x1","y"])
```

**Grupo 1**

```python
X1 = df[["x1"]].iloc[0:6]
y1 = df["y"].iloc[0:6]

model1 = DecisionTreeClassifier(max_depth=1)
model1.fit(X1,y1)
model1.predict(df[["x1"]].iloc[6:9])
```

```python
tree.plot_tree(model1,class_names=["no","si"])
```

**Grupo 2**

```python
X2 = df[["x1"]].iloc[3:9]
y2 = df["y"].iloc[3:9]

model2 = DecisionTreeClassifier(max_depth=1)
model2.fit(X2,y2)
model2.predict(df[["x1"]].iloc[0:3])
```

```python
tree.plot_tree(model2,class_names=["no","si"])
```

**Grupo 3**


```python
X3 = pd.concat([df[["x1"]].iloc[0:3],df[["x1"]].iloc[6:9]])
y3 = pd.concat([df[["y"]].iloc[0:3],df[["y"]].iloc[6:9]])

model3 = DecisionTreeClassifier(max_depth=1)
model3.fit(X3,y3)
model3.predict(df[["x1"]].iloc[3:6])
```

```python
tree.plot_tree(model3,class_names=["no","si"])
```

**Accuracy de cada grupo**

```python
print("Ac train1: ",model1.score(X1,y1))
print("Ac test1: ",model1.score(df[["x1"]].iloc[6:9], df["y"].iloc[6:9]))
```

```python
print("Ac train2: ",model2.score(X2,y2))
print("Ac test2: ",model2.score(df[["x1"]].iloc[0:3], df["y"].iloc[0:3]))
```

```python
print("Ac train3: ",model3.score(X3,y3))
print("Ac test3: ",model3.score(df[["x1"]].iloc[3:6], df["y"].iloc[3:6]))
```
