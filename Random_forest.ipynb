```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
clf = RandomForestClassifier(n_estimators=3, max_depth=2)
```

```python
clf.fit(X, y)
```

```python
muestra = X[0:1, :]
```

```python
clf.predict(muestra)
```

```python
clf.predict_proba(muestra)
```

```python
clf.estimators_
```

```python
proba1 = clf.estimators_[0].predict_proba(muestra)
proba2 = clf.estimators_[1].predict_proba(muestra)
proba3 = clf.estimators_[2].predict_proba(muestra)
```

```python
print("Probabilidad de la muestra para el primer árbol:", proba1)
print("Probabilidad de la muestra para el segundo árbol:", proba2)
print("Probabilidad de la muestra para el tercer árbol:", proba3)
```

```python
plt.figure(figsize=(10, 10))
plot_tree(clf.estimators_[0], filled=True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plot_tree(clf.estimators_[1], filled=True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plot_tree(clf.estimators_[2], filled=True)
plt.show()
```

```python
prom_total=(proba1+proba2+proba3)/3
prom_total
```
