```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(X, thetas):
    z = np.dot(X, thetas)
    return sigmoid(z)

def cost_function(X, y, thetas):
    n = len(y)
    y_pred = predict(X, thetas)
    cost = -(1/n) * np.sum(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
    return cost

def gradient_descent(X, y, thetas, learning_rate, iterations):
    n = len(y)
    costs = []
    for i in range(iterations):
        y_pred = predict(X, thetas)
        gradient = (1/n) * np.dot(X.T, (y_pred - y))
        thetas -= learning_rate * gradient
        cost = cost_function(X, y, thetas)
        costs.append(cost)
    return thetas, costs


np.random.seed(123)
X = np.random.random(size=(12, 2))*5
X[9, :]= [1,4]
y = np.array([0, 2, 0, 1, 0, 2, 0, 2, 0, 2, 1, 1])


X = np.insert(X, 0, 1, axis=1)


y_labels = [np.where(y == i, 1, 0) for i in range(3)]


thetas = np.zeros((3, 3))


for i in range(3):
    thetas[i], costs = gradient_descent(X, y_labels[i], thetas[i], learning_rate=0.1, iterations=1000)


x1 = np.linspace(0, 5)
for i in range(3):
    plt.scatter(X[:, 1][y == i], X[:, 2][y == i])
    x2 = -(thetas[i][0] + thetas[i][1]*x1) / thetas[i][2]
    print("Theta1: ",thetas[i][1]," theta 2: ",thetas[i][2]," theta 0: ",thetas[i][0])
    plt.plot(x1, x2)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```
